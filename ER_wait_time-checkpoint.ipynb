```python
import pandas as pd
import matplotlib.pyplot as plt
file_path = "EMS_Incident_Dispatch_Data.csv"  
df = pd.read_csv(file_path, nrows=1_000_000)
print(df.columns)
df.head()
```

```python

df['INCIDENT_DATETIME'] = pd.to_datetime(df['INCIDENT_DATETIME'])

# drop rows with  0 or NaN
df = df[df['DISPATCH_RESPONSE_SECONDS_QY'].notna()]
df = df[df['DISPATCH_RESPONSE_SECONDS_QY'] > 0]

# get daily average response time
df['date'] = df['INCIDENT_DATETIME'].dt.date
daily_df = df.groupby('date')['DISPATCH_RESPONSE_SECONDS_QY'].mean().reset_index()
daily_df.columns = ['date', 'avg_dispatch_response_time_sec']

# time series plot
plt.figure(figsize=(12, 5))
plt.plot(daily_df['date'], daily_df['avg_dispatch_response_time_sec'], label='Daily Avg Response Time')
plt.title("Daily Average EMS Dispatch Response Time")
plt.xlabel("Date")
plt.ylabel("Seconds")
plt.grid(True)
plt.tight_layout()
plt.show()

#plot
daily_df['date'] = pd.to_datetime(daily_df['date'])
filtered_df = daily_df[(daily_df['date'] >= "2018-01-01") & (daily_df['date'] <= "2024-12-31")].copy()
plt.figure(figsize=(12, 5))
plt.plot(filtered_df['date'], filtered_df['avg_dispatch_response_time_sec'], label='Daily Avg Response Time')
plt.title("Filtered: Daily Avg EMS Dispatch Response Time (2018â€“2024)")
plt.xlabel("Date")
plt.ylabel("Seconds")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python

df_feat = filtered_df.copy()

# time features
df_feat['day_of_week'] = df_feat['date'].dt.dayofweek              # 0 = Monday
df_feat['is_weekend'] = (df_feat['day_of_week'] >= 5).astype(int)  # 1 if Sat/Sun, else 0

# lag features
df_feat['lag_1'] = df_feat['avg_dispatch_response_time_sec'].shift(1)
df_feat['lag_7'] = df_feat['avg_dispatch_response_time_sec'].shift(7)

# rolling averages
df_feat['rolling_7'] = df_feat['avg_dispatch_response_time_sec'].rolling(window=7).mean()
df_feat['rolling_14'] = df_feat['avg_dispatch_response_time_sec'].rolling(window=14).mean()

# drop None values 
df_feat.dropna(inplace=True)
df_feat.head()
```

```python

df_feat['date'] = pd.to_datetime(df_feat['date'])

# split  time
train_df = df_feat[df_feat['date'] < '2023-01-01']
test_df = df_feat[df_feat['date'] >= '2023-01-01']

features = ['day_of_week', 'is_weekend', 'lag_1', 'lag_7', 'rolling_7', 'rolling_14']
target = 'avg_dispatch_response_time_sec'
X_train = train_df[features]
y_train = train_df[target]
X_test = test_df[features]
y_test = test_df[target]
!pip install xgboost
```

```python
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np

# training
model = XGBRegressor(n_estimators=100, learning_rate=0.1, random_state=42)
model.fit(X_train, y_train)

# test set
y_pred = model.predict(X_test)

rmse = np.sqrt(mean_squared_error(y_test, y_pred))
mae = mean_absolute_error(y_test, y_pred)

print(f"Test RMSE: {rmse:.2f} seconds")
print(f"Test MAE: {mae:.2f} seconds")
```

```python
import matplotlib.pyplot as plt

results_df = test_df.copy()
results_df['predicted'] = y_pred

plt.figure(figsize=(14, 5))
plt.plot(results_df['date'], results_df['avg_dispatch_response_time_sec'], label='Actual', linewidth=2)
plt.plot(results_df['date'], results_df['predicted'], label='Predicted', linestyle='--')
plt.title('EMS Dispatch Response Time: Actual vs Predicted (2023â€“2024)')
plt.xlabel('Date')
plt.ylabel('Seconds')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
